```python
### 1 move file to subdirectories and update recap
import os
import shutil
import csv

os.chdir(r'/Users/name/Downloads/FileOrganizer/files')

def subfolder(file): #define function to use in the movement of files
    if file.endswith(('txt','odt')):
        return('doc') 
    if file.endswith(('mp3')):
        return('audio')
    if file.endswith(('png', 'jpeg', 'jpg', 'PNG')):
        return('image')

#check if the file recap already exists. If it doesn't exists, the programme creates new one
if not os.path.exists('recap1.csv'):
    with open('recap1.csv', 'w+', newline='') as f:
        wr = csv.writer(f)
        # Write a header row
        wr.writerow(['Filename', "size"])
        print('RECAP1.CSV JUST CREATED')
else:
    print('RECAP1.CSV ALREADY EXISTS')


# move file from folder to subfolders
pdc = r'../files' #parent directory 
folders = ['audio','doc','image'] # CREATE NEW SUBFOLDERS "audio, image, docs"
for folder in folders:
    if not os.path.exists(os.path.join(pdc,folder)):
        os.mkdir(os.path.join(pdc,folder))  

for root, dirs, files in os.walk('.'): 
    for file in files:
        with open('recap1.csv', 'a+', newline='') as f: #update recap with the moved files 
            wr = csv.writer(f)
            num_files = 0   
            path = os.path.join(root, file) 
            fn,fe = os.path.splitext(file)
            size = os.stat(path).st_size
            
            if not os.path.exists(file):
                #print the files in the directory
                if file.endswith(('txt','odt')):
                    print(fn,':doc ;', "size:",size, "B")
                elif file.endswith(('mp3')):
                    print(fn, ':audio ;', "size:",size, "B")
                elif file.endswith(('png', 'jpeg', 'jpg', 'PNG')):
                    print(fn, ':image ;', "size:",size, "B")             
            if os.path.exists(file):
                num_files += 1
                if fe != '.csv':
                    wr.writerow((fn, size)) # fill the columns 'filename' and 'size' with the file moved                
                if subfolder(file) == 'doc': #use the function previously created to move the file through directories
                    shutil.move(file, os.path.join(pdc,'doc'))                   
                elif subfolder(file) == 'audio':
                    shutil.move(file,os.path.join(pdc,'audio')) 
                elif subfolder(file) == 'image':
                    shutil.move(file,os.path.join(pdc,'image'))
```

```python
# user-friendly program to move files towards subdirectories
import argparse

# create the variable parser
parser = argparse.ArgumentParser()

#parse infile as mandatory (positional) argument and directory as optional arg
parser.add_argument('infile', type=str, 
                    help= "write the filename and extension to move (ex: trump.jpeg) ")
parser.add_argument('directory', type=str, help="write the destination path")

# create the variable to use only when call the function move_files
args = parser.parse_args()

#define function to move files in subfolders
def move_file(file, directory):
    if os.path.isfile(file):
        shutil.move(file, directory)
        #return the sentence and import the name and size of the file called, in csv
        return print('Well done, the program has just moved the file. CSV recap is updated'), csv_(file)
    elif not os.path.isfile(file):
        #if the file is not in the directory 'files', will return this sentence
        return print('Ops, the file is not present in the current directory') 
    
#create the csv file while the first file is parsed with the CLI    
try:
    with open("recap2.csv", "x", newline='') as f: #create new csv file
        wr= csv.writer(f)
        wr.writerow(['Filename', 'Size'])
        print('CSV recap has just been created')
except FileExistsError:
    #if CSV has been already created, when call the mandatory argument, will print the following
    print('CSV recap has been already created') 
    

#define function to parse the files and organize them into csv file 
def csv_(filename): #this function will be returned in the move_file function after shutil
    with open('recap2.csv', 'a+', newline='') as f:
        wr = csv.writer(f)
        num_files = 0
        for root, dirs, files in os.walk('.'): 
            for file in files:
                if file == filename:
                    num_files += 1
                    path = os.path.join(root, file)
                    fn,fe = os.path.splitext(file)
                    size = os.stat(path).st_size                        
                    move = wr.writerow((fn,size)) 
                    return move
                
#call the function move_file and parsing the arguments
move_file(args.infile, args.directory)
```
